```python
#Step 1---Taking Data and Creating Dataframe
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/diazoniclabs/Machine-Learning-using-sklearn/master/Datasets/news.csv')
df
```

```python
#step4---Assigning input and output
x=df.iloc[:,0].values
x
```

```python
y=df.iloc[:,2].values
y
```

```python
#Step5---Assigning Train and Test Variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
#step6---APPLYING TF IDF VECTORIZER
from sklearn.feature_extraction.text import TfidfVectorizer
vect=TfidfVectorizer()
x_train_v=vect.fit_transform(x_train)
x_test_v=vect.transform(x_test)
```

```python
#step7---APPLYING CLASSIFIER
from sklearn.svm import SVC
model=SVC()
```

```python
#step8---FIT THE MODEL
model.fit(x_train_v,y_train)
```

```python
#step9---PREDICT THE OUTPUT
y_pred=model.predict(x_test_v)
```

```python
y_pred
```

```python
#step10---Evaluation Accuracy Score
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
#PIPELINING
from sklearn.pipeline import make_pipeline
text_model = make_pipeline(TfidfVectorizer(),SVC())
text_model.fit(x_train,y_train)
```

```python
#predictor varibale
y_pred1 = text_model.predict(x_test)
y_pred1 # these are predicted outputs for pipelined model
```

```python
y_test #actual output
```

```python
#To check the accuracy of the pipelined model
accuracy_score(y_pred1,y_test)*100
```
